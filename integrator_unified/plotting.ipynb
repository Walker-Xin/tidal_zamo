```python
names = ["Q", "a", "force"]
data = pd.read_csv("data/data_a_pro.dat", sep=",", header=None, names=names)
# Create a pivot table to reshape the data for plotting
pivot_table = data.pivot(index='a', columns='Q', values='force')
```

```python
# Create the plot
fig, ax = plt.subplots(figsize=size_square)
c = ax.pcolormesh(pivot_table.columns, pivot_table.index, pivot_table.values, cmap=cmap)

# Add color bar
fig.colorbar(c, ax=ax)

#

# Set labels
ax.set_xlabel('Q')
ax.set_ylabel('a')
ax.set_title('Colormap of Force')

plt.show()
```
